```python
# -------------CPLEX Reformulation ------------------
def cplex_linearized( K,J ,  R_K, BUDGETS, S_K  , **kwargs):  
     
    #iterations +=1 
    #     ----- Result Variable
#     K = len(BUDGETS)
#     J = len(BUDGETS)
    Cust_Ass =[]              #Empty list to store decision variable x_i_j
    Prices =[]
    Q  =  []
    lower_objective = 0
    max_b = max(BUDGETS)
    mdl = Model(name='lowerlevel' )
    L = min(BUDGETS) 
    U = max(BUDGETS) 
    pi_min = min(min(R_K))
    pi_max = max(BUDGETS)
    epsilon=0.01
    weight = 0.8
    big_M = 1000
   
    J_range = range(0,J )
    K_range = range(0,K)
    # ------ decision variables ---
    x  = mdl.binary_var_matrix(K,J ,"cust_assign", key_format ="_%s")
    p  = mdl.continuous_var_matrix(K,J, name="price"  , key_format ="_%s")
    pi = mdl.integer_var_list( J , name="pi" , key_format ="_%s")
    q  = mdl.binary_var_matrix(K,J ,"aux_q", key_format ="_%s")
    t  = mdl.continuous_var_matrix(K,J ,name="aux_t", key_format ="_%s")

    # --------constraints 1---
    for k in K_range:  
        mdl.add_constraint( mdl.sum(  x[k,j] for j in J_range if S_K[k][j] > 0 )  <= 1   )
    
    # --------constraints 2---
#      Binary constraints 

    # --------- constraints3 ---
    for k in K_range:
        for j in J_range:
            for i in J_range:
                if (i !=j  and  S_K[k][j] > 0 and S_K[k][i] > 0    ):
                    mdl.add_constraint(  ( weight*S_K[k][j] +  (1-weight) *R_K[k][j] ) *x[k,j ] - ( 1-weight)*p[k,j]  >= ( weight*S_K[k][i] + (1-weight)*R_K[k][i] )  *x[k,j] - (1-weight) * pi[i] - big_M * q[k,i]  )  
   
    # --------- constraints4 ---
    for k in K_range:
        for j in J_range:
            if S_K[k][j] > 0:
                mdl.add_constraint( p[k,j]  <= pi[j] )      
            
    # --------- constraints5 ---     
    for k in K_range:
        for j in J_range:
            if S_K[k][j] > 0:
                mdl.add_constraint( p[k,j]  <= BUDGETS[k] * x[k,j] )
    
     # --------- constraint redundant ---   FOR PROOF THIS CONSTRAINT BECOMES REDUNDANT   
#     for k in R:
#         if S_K[k][j] > 0:
#               mdl.add_constraint( mdl.sum(   p[k,j]   for j in R  )  <= BUDGETS[k]    )
                 
    # --------- constraints6 ---
    for k in K_range:
        for j in J_range:
            if S_K[k][j] > 0:
                mdl.add_constraint( p[k,j]  >= pi[j] - max_b * (1-  x[k,j] )  )
        
#     # --------- constraints7 ------------  Auxiliary q  
    for k in range(0,K):
        for i in range(0,J):
            if S_K[k][i] > 0:
                mdl.add_constraint( BUDGETS[k]  >=  pi[i] -   t[k,i] + L * q[k,i]  )
    
    # --------- constraints8 ------------      
    for k in range(0,K):
        for i in range(0,J):
            if S_K[k][i] > 0:
                mdl.add_constraint(  BUDGETS[k]   <=  t[k,i] - epsilon * q[k,i]  + U*(1- q[k,i] )    )  
                
    # --------- constraints9 a------------      <=   pi_max * q[k,i] 
    for k in range(0,K):
        for i in range(0,J):
            if S_K[k][i] > 0:
                mdl.add_constraint( pi_min* q[k,i] <= t[k,i]   )  
    
    # --------- constraints9 b ------------     
    for k in range(0,K):
        for i in range(0,J):
            if S_K[k][i] > 0:
                mdl.add_constraint(   t[k,i]   <=   pi_max * q[k,i]  )  
    
#     --------- constraints10_a ------------      <=   -pi_min * (1- q[k,i])
    for k in range(0,K):
        for i in range(0,J):
            if S_K[k][i] > 0:
                mdl.add_constraint( -pi_max* ( 1- q[k,i]) <= t[k,i] - pi[i]   )   
    
   # --------- constraints10 b ------------      
    for k in range(0,K):
        for i in range(0,J):
            if S_K[k][i] > 0:
                mdl.add_constraint(   t[k,i] - pi[i] <=   -pi_min * (1- q[k,i])   )   
                
    # --------- constraints11 ------------
    for k in K_range:
        for j in J_range :
            if S_K[k][j] > 0:
                mdl.add_constraint( p[k,j] >= 0   )   
                
    # --------- constraints11 ------------                                     
    for j in J_range:
        if S_K[k][j] > 0:
            mdl.add_constraint( pi[j]    >= 0   )               
    
    # ------------ objective ---         
    mdl.maximize( mdl.sum(   p[k,j]   for k in range(0,K) for j in range(0,J) if S_K[k][j] > 0  ) )        
    
    # ------------ KPI  ---   
    mdl.add_kpi(  mdl.sum( x[k,j ]*pi[j]      for k in K_range for j in J_range )  , publish_name = 'UpperObjective')  

    # ------------ KPI Lower Level ---   
    mdl.add_kpi( mdl.sum(   weight*  x[k,j ] *S_K[k][j] +  x[k,j] *(1-weight)* ( R_K[k][j] - pi[j]  ) for k in range(0,K) for j in range(0,J)) , publish_name = 'LowerObjective') 
    
    
    #------------SOLVE SETTINGS --------------------------------
    mdl.parameters.mip.tolerances.mipgap.set(0.00000000000000001)          #Assigns Gaps 
    mdl.set_time_limit(14400)                    #Setting Time Limit 
    mdl.parameters.mip.strategy.file = 3         # Tranferring Node FIle , mmeory full issue 
#     model.context.cplex_parameters.threads = 4   # Number of Threads . NOT USEDDD 

    print('NUMBER OF THREADS BEING USED , DISPLAYING BEFORE SOLVE ' , mdl.context.cplex_parameters.threads.get()   )
    msol = mdl.solve(log_output=True)                 #log_output=True              # msol is a data structure of the class Solve Solution
    
    # ----- Retrieving values of decision variables Assignment 
    for i in range(0,K):
            for j in range(0,J):
                    ld = msol.get_value(x[i,j ])
                    Cust_Ass.append(ld) 
    # ----- Retrieving values of decision variables PRICES
    for j in range(0,J):
        ld = msol.get_value(pi[ j ])
        Prices.append(ld) 
    
    # ----- Retrieving values of q
    for k in range(0,K):
            for i in range(0,J):
                    ld = msol.get_value( q[k,i] )
                    Q.append(ld) 
    
    #----------------Solve module , solve detals ----------------
    print('Solve Details in order mip relative gap, best bound' ,msol.solve_details.mip_relative_gap , msol.solve_details.best_bound , msol.solve_details.status )
    print('\n Gap is ', msol.solve_details.gap)
    
    # Some Prints 
    mdl.export_as_lp(basename="cplex_%s", path="D:\AA MASC\C Codes\FINAL" )  #Printing the constraints 
    print("time limit = ",mdl.get_time_limit()) #Prints time limit 
    print("threads = ",mdl.parameters.threads.get()) #Prints thread
    print('threads  ' , mdl.context.cplex_parameters.threads.get()   ) 
#     time_lim =  mdl.get_time_limit()           #getting the time limit 
#     mdl.report()                  #Prints KPIs  , value of obj
#     mdl.print_solution() 
    upper_objective = msol.get_objective_value()                # Retrieves obj value 
    upper_objective_kpi = msol.kpi_value_by_name('UpperObjective')
    lower_objective_kpi = msol.kpi_value_by_name('LowerObjective')
    
    return Cust_Ass, Prices  ,upper_objective ,upper_objective_kpi, lower_objective_kpi , Q
```

```python
# ----------Funtion to Compute , Assignment 
def prod_number(cust_list, no_cust, no_prod  ):
    numbered_list = []
    iterator = 0
    for i in range(no_cust):
        flag = 0
        j = 0
        for j in range(no_prod ):  
#             print(int(str(i) + str(j)))
#             if ( cust_list[int(str(i) + str(j))]  >  0 ) :  
            if ( cust_list[iterator]  >  0 ) :  
                numbered_list.append(j+1)
                flag=1
            j += 1
            iterator += 1
        if flag ==0:
            numbered_list.append(0)
            
    return numbered_list

# This list contains actual index i.e. Product1, Prod 2 NOT python indices from 0,
# A value of 0 indicates customer did not made any purchase 
```

```python
# ----------Funtion to Check Optimality of Lower Level 
# -------------CPLEX LOWER LEVEL ------------------
def master_lower_level( K, J , PRICES, R_K, BUDGETS, S_K  , **kwargs):  
   
     
    #iterations +=1 
    #     ----- Result Variable (Variables to store) ----------------
    Cust_Ass =[]
    lower_objective = 0
    
    mdl = Model(name='lowerlevel' )

    weight =0.8          # Weight 
   
    R = range(0,K)
    J_Range = range(0,J )
    # ------ decision variables ---
    x  = mdl.binary_var_matrix(K,J ,"cust_assign", key_format ="_%s")
     
    # --------constraints 1---
    for i in R:  
        mdl.add_constraint( mdl.sum(  x[i,j] for j in J_Range )  <= 1   )

    # --------- constraints2 ---
    for i in R:
        mdl.add_constraint( mdl.sum(   x[i,j]* PRICES[j] for j in J_Range  )  <= BUDGETS[ i]    )
      
    # ------------ objective ---         
    mdl.maximize( mdl.sum(   weight*  x[i,j ] *S_K[i][j] +  x[i,j ] *(1-weight)* (R_K[i][j] - PRICES[j]  )     for i in R for j in J_Range) )      
    
    # ------------ KPI  ---   
    mdl.add_kpi(  mdl.sum( x[i,j ]*PRICES[j]      for i in R for j in J_Range )  , publish_name = 'UpperObjective')  

 
    msol = mdl.solve()                 #log_output=True              # msol is a data structure of the class Solve Solution
    # ----- Retrieving values of decision variables 
    for i in range(0,K):
            for j in J_Range:
                    ld = msol.get_value(x[i,j ])
                    Cust_Ass.append(ld) 
                    
    lower_objective = msol.get_objective_value()                # Retrieves obj value 
    upper_objective = msol.kpi_value_by_name('UpperObjective')
    
    return Cust_Ass , lower_objective ,upper_objective

## VALIDATING PRICES FROM CPLEX 
soln = master_lower_level(K, J ,  Opt_Prices, R_K, BUDGETS, S_K  ) 
print('Numbered Prod Purchased by checking   \t ',  prod_number(soln[0], K , J ))
print('Result Obtained from Reformulation   \t', prod_number(Cust_Ass,K,J) )
if (soln[0] == Cust_Ass):
#     print(soln[0] , Cust_Ass)
    print('Lower Level is Optimal ')
else:
#     print(soln[0] , Cust_Ass)
    print('Not Optimal')
```
